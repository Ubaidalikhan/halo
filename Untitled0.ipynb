```python
import numpy as np
from sklearn.linear_model import Perceptron

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,1])

model = Perceptron()

model.fit(X,y)

prediction = model.predict([[0,0]])
print(" for [0,0]:", prediction)

prediction = model.predict([[0,1]])
print(" for [0,1]:", prediction)

prediction = model.predict([[1,0]])
print(" for [1,0]:", prediction)

prediction = model.predict([[1,1]])
print(" for [1,1]:", prediction)
```
